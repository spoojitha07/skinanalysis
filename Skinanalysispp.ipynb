```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
df = pd.read_csv('cleaned_skin_analysis.csv')
df
```

```python
sns.countplot(data=df, x = 'Skin_Type')
plt.title('Distributions of different Skin_type')
plt.xticks(rotation = 45)
plt.show()
```

```python
concerns = Counter([item for sublist in df['Skin_Concerns'].apply(eval) for item in sublist])
pd.Series(concerns).sort_values(ascending=False).plot(kind='bar')
plt.title('Most Common Skin Concerns according to data')
plt.show()
```

```python
products = Counter([item for sublist in df['Products_Used'] .apply(eval) for item in sublist])
pd.Series(products).sort_values(ascending=False).plot(kind='pie',autopct='%1.1f%%')
plt.title('Most Products are used')
plt.show()
```

```python
brands = Counter([item for sublist in df['Brand_Preference'] .apply(eval) for item in sublist])
pd.Series(brands).sort_values(ascending=False).plot(kind='pie',autopct='%1.1f%%')
plt.title('Most Brands are prefered')
plt.show()
```

```python
sns.countplot(data=df, x="Gender", hue="Skin_Type", palette="pastel")
plt.title("Skin Type by Gender")
plt.legend(title="Skin Type")
plt.tight_layout()
plt.show()
```
